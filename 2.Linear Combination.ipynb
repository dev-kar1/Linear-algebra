# Linear combination (LC)
Let $V$ be a vector space over field $F$

also let $v_1,v_2,\cdots,v_m\in V$ and $a_1,a_2,\cdots,a_m\in F$ then,

$v=a_1 v_1+a_2 v_2+\cdots+a_m v_m$  is called a linear combination.
- More precisely we call $v$ is linear combination of vectors $v_1,v_2,\cdots v_m$.

For example,
$v_1=(1,3),v_2=(1,1)$ and $a_1=2,a_2=4$ then the respective linear combination is: $$v=a_1 v_1+a_2+v_2=2\cdot(1,3)+4\cdot(1,1)=(2,6)+(4,4)=(6,10)$$

- Moreover we would like to use the vector addition and scalar multiplication together and term it as a linear combination.
- we already know some lc such as $v_1+v_2,v_1-v_2,etc.
$

Theres a lot to learn from the picture above
- $v_1$ being a vector we can do a lc of $v_1$ with scalar $-1$ to get vector $-v_1$
- we can take lc of $v_1,v_2$ with scalar $1$ to obtain vector $v_1+v_2$. The process shown here to obtain the same is well known as triangle law of vector addtion.
- More such things can be understood drom this picture, it is encoragd to the reader to extract some sense from the geometry.

```python
#implementing linear combinations in python
a_1,a_2=2,4 #defining scalars
# defining vectors
v_1,v_2=[1,3],[1,1]
w_1,w_2=[0,0],[0,0]
#initializing the vecor (0,0) to store lc vector
linear_combination=[0,0]
for i in range (2):
    w_1[i]=a_1*v_1[i]
    w_2[i]=a_2*v_2[i]
    linear_combination[i]=w_1[i]+w_2[i]
print(linear_combination)
```

# Vector Span
Let $v_1,v_2,\cdots v_m$ be $m$-vectors then the span is defined as set of all possible linear combinations of the vectors with the scalars from the field.

**For example** the span of $\{(1,0),(0,1)\}$ is $\mathbb{R}^2$

# Subaspace
Let $V$ be a vector space, we say $U$ is a subspace if:
1. $\vec{o} \in u$
2. For any vectors $x,y \in U$ we have $x+y\in U$
3. for any scalar $a$ and vector $x\in U$ we have $ax\in U$ 

- Subspaces such as null space, row space, column space, range space, etc. would be in frequent use in linear algebra.

# Linear dependance and independance
A set of vectors $v_1,v_2,\cdots v_m$ are said to be linearly dependent if there exists scalars $a_1,a_2,\cdots a_m$ not all zero such that $a_1 v_1+a_2 v_2+\cdots +a_m v_m=0$

# Basis
A set of linearly independent vectors is called a basis if the set of all linear combinations generates the vector space.
- The number of elements a basis holds is called the dimention of the vectorspace.
