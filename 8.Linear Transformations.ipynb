# Linear Transformation
- Also known as linear functional, linear maps.
- It is a rule that takes vectors from one space to another space.
- Mathematically, if $U$ and $V$ are vector spaces and $T:U\to V$ is a linear map if the following holds:
1. $T(0_U)=0_V$
2. $T(x+y)=T(x)+T(y)$
3. $T(\alpha x)=\alpha T(x)$
- $U$ is called the domain space(elements are called pre-images) and $V$ is the range space(elements are called images).
- T maps the zero vector in $U$ to the zero vector in $V$.
- The sum of images of x an y is same as the image of sum of x and y.
- The process of streching the image is same as stretching its pre image and then transforming it.
- Linear transformation is same as matrix multiplication. $T(x)=Ax$
- Instead of giving a rule, it is ok if we get a matrix to play with.

```python
import numpy as np
x=np.array([[-1],[0],[1]])
y=np.array([[-3],[0],[2]])
o=np.array([[0],[0],[0]])
T = np.array([[-1,0,0],[0,-1,0],[0,0,-1]])
# property 1
a=T@o
print('T(0)=',a)
b=T@x
c=T@y
lhs=b+c
rhs=T@(x+y)
print('LHS=',lhs)
print('RHS=',rhs)
# sa,e treatment can be done by picking some aplha
```

- The matrix taken in previous example corresponds to the negation transformation defined as
$$T(x,y,z)=(-x,-y,-z)$$
- **Reversal transform** is defined as $T(x,y,z)=(z,y,x)$
$$\text{Reversal matrix is: }\begin{bmatrix}0&0&1\\0&1&0\\1&0&0\end{bmatrix}$$
- **Zero transform** is defined as $T(x,y,z)=(0,0,0)$

## Translation and affine mapping
- **Translation** is the processs of shifting the vectors in a space with a specifc distance. For instance, $T(x,y)=(x+1,y+1)$ is a translation by 1 units.
- **Note:** Translation is not linea transformation as it fails to fix $0$ vector.
## Affine mapping
It is a map defined as:
$$M(x)=T(x)+b$$
or $$M(x)=Ax+b$$
whcih is a combination of a linear transformation and a translation.
- $b$ is called the translation vector.


## Affine combination
- It is just same as linear conditiion with a restrictions to the coefficients that their sum is 1.
- In other words we say $v_1,v_2$ form an affine combination if for scalars $a,b$ we have $$av_1+bv_2\\\text{ such that }a+b=1$$
- More precisely let $v_1,v_2,\cdots v_m$ are vectors and we have scalars $a_1,a_2,\cdots a_m$ with $\sum_{n=1}^{m}a_n=1$ then $v$ is an affine combination.
$$v=\sum_{n=1}^{m}a_nv_n$$
- Set of all possible affine combinations of a set is called the **affine span**

# Special linear mappings
## Scaling
- It is the process of changing length, physically can be thought as zooming in and out or streching and shrinking.
- $T(x)=\alpha x$, $\alpha$ is called the scaling factor.
- If $\alpha >1$ then vector streaches i.e zooms in.
- If $0<\alpha <1$ then vector shrinks i.e zooms out.
- If $\alpha <0$ then vector shrinks and streaches in the opposite direction.
## Reflexion
- It is the process of obtaining mirror image of a vector with respect to a reference.
- For example $T(x,y)=(-x,-y)$ is the reflection of vector w.r.t x-axis in $\mathbb{R}^2$
- The usually known orthogonal matrix(A) acts as an reflection of any vector w.r.t a line passing through origin and angle of inclination $\theta$ is given by:
$$\begin{pmatrix}\cos{2\theta}&\sin{2\theta}\\ \sin{2\theta}&-\cos{2\theta}\end{pmatrix}$$
- For reflection around x axis $\begin{pmatrix}1&0\\0&-1\end{pmatrix}$
- For reflection around line 45 degrees from x axis $\begin{pmatrix}0&1\\1&0\end{pmatrix}$
- For reflection around line 90 degrees from x axis $\begin{pmatrix}-1&0\\0&1\end{pmatrix}$
- For particular reflection we cannot rely on numpy and have to write these matrices by our own.

## Rotation
- **Rotation** is the process of rotating a vector by some angles. For instance, $T(x,0)=(0,x)$ is a roation by  $90$ degrees.
- It is always meant to be anti-clockwise rotation.
- The matrix for rotation is $$\begin{pmatrix}\cos{\theta}&-\sin{\theta}\\ \sin{\theta}&\cos{\theta}\end{pmatrix}$$
- Rotation we can define for different values of $\theta$.

## Projection
- It is defined as a transformation $T$ such that $T^2=T$.
- Mostly used to change a higher dimentional space to a lower dimention.
- For example $T(x,y)=(x,0)$ is the projection of xy plane on x axis.

```python
# scaling a vector [1,2,3] be scaling factor a=5
a=[1,2,3]
for i in range(len(a)):
    a[i]=5*a[i]
print(a)
# using numpy
import numpy as np
alpha=5
x=np.array([[1],[2],[3]])
print(alpha*x)
```

# More on Projection maps
- These are much useful in performing dimentionality reduction.
- Dimentionality reduction is a technique of reducing size of vectors to make calculations computationally efficient.
- Linear regression is also a projection of vectors.
- As discussed earlier projection as a transformation can be defined as a matrix multiplication.
## Projection onto lines
- The projection of a vector onto a line is the shadow of vector w.r.t the line.
- Projection of $v_1$ on $v_2$ is given by $\frac{<v_1,v_2>}{||v_2||}v_2$
