# Inner Product
The dot product or inner product of two vectors is defined as:
    
$$ \begin{align}
<x,y>:=x^T y=\
     \begin{bmatrix}
           x_{1} \\
           x_{2} \\
           \vdots \\
           x_{m}
         \end{bmatrix} ^T
         \begin{bmatrix}
           y_{1} \\
           y_{2} \\
           \vdots \\
           y_{m}
         \end{bmatrix} 
         = \begin{bmatrix}
           x_{1} ~x_{2} \cdots ~x_{m}
         \end{bmatrix}\begin{bmatrix}
           y_{1} \\
           y_{2} \\
           \vdots \\
           y_{m}
         \end{bmatrix}
  \end{align} =x_1 y_1+x_2 y_2 +\cdots x_m y_m$$
- Inner product is a specific dot product in linear algebra

# Dot product
The dot product of two vectors is defined as:
    
$$ \begin{align}
xy:=x^T y=\
     \begin{bmatrix}
           1\\
           2 \\
           4
         \end{bmatrix} ^T
         \begin{bmatrix}
           4 \\
           1 \\
           2
         \end{bmatrix} 
         = \begin{bmatrix}
           1 ~2 ~4
         \end{bmatrix}\begin{bmatrix}
           4 \\
           1 \\
           2
         \end{bmatrix}
  \end{align} =1\times 4+2\times 1+ 4 \times 2=4+2+8=14$$

```python
#Finding inner product of vectors
x,y,z=[1,2,4],[4,1,2],[0,0,0] #vector inputs
for i in range (3):
    z[i]=x[i]*y[i] #multiplying vectors component wise and storing them in vector z
    #a=z[i]
   # print(a)
sum(z) #sum() function gives the sum of all stored values in array
```

# Norm
The norm of a vector is the length between terminal points. 
It is denoted as $||~||$ 

Norm satisfies the following properties:
1) Absolutely Homogeneous
$$||\alpha x||=|\alpha| ||x||$$
2) Triangle Inequality
$$||x+y||\le||x||+||y||$$
3) Positive Definite $$||x||\ge 0\text{ and } ||x||=0\text{ iff } x=0$$ 
We would define the most commonly used norm in this section. But any function as a norm can be defined and used as long as the three properties hold.

1. **Euclidean Norm** is denoted as $||x||_2$
Norm can be defined as the square root of inner product.

$$||x||_2=\sqrt{<x,x>}$$

Or we can write:
    $$||x||_2=\sqrt{x_1^ 2 +x_2 ^2+\cdots +x_m ^2}$$

```python
# Euclidean norm of the vector (1,1,2)
a,x=[0,0,0],[1,1,2] 
for i in range (3):
    a[i]=x[i]*x[i] #storing the square values to add later 
    #print(a)
#print(sum(a)) # sum() finds the sum of elements in list, we want to find its squareroot
b=sum(a) # storing value to find its squareroot
import math #importing math libarary to find sqrt
c=math.sqrt(b)
print(c)
```

```python
# Using numpy
import numpy as np
x=np.array([1,1,2])
a=np.inner(x,x)
import math
print(math.sqrt(a))
```

2. **Taxi cab/Manhattan norm**
    is defined as the sum of modulus of each coordinates and denoted as $||x||_1$

$$||x||_1=\sum_{i=1}^{n}|x_i |  $$

For example Manhattan norm of the vector $(1,1,2)$ is 
$$||x||_1=|1|+|1|+|2|=4 $$


```python
# Manhattan norm of the vector (1,1,2)
x=[1,1,2] 
print(sum(x)) # sum() finds the sum of elements in list
```

```python
x=np.array([1,1,2])
a=np.linalg.norm(x,1) #1-gives the taxicab norm
b=np.linalg.norm(x,2) #2-gives the euclidean norm
print("Taxi cab norm:",a)
print("Euclidean norm:",b)
```

3. **Max norm/ infinity norm** is defined as the maxium of all the values of the coordinates of the vector and then taking its modulus. It is denoted as $||x||_{\infty}$

For (1,1,2) the max norm is:
$$||(1,1,2)||_{\infty}=Max\{|1|,|1|,|2|\}=2$$
